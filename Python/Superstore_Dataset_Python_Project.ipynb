## Data Cleaning

This section checks for missing values, duplicate records, and converts data types where necessary.

```python
# Convert 'Order Date' Datatype into datetime
Superstore['Order Date'] = pd.to_datetime(
    Superstore['Order Date'],
    format='mixed')
```

```python
# Convert 'Ship Date'Datatype into datetime
Superstore['Ship Date'] = pd.to_datetime(
    Superstore['Ship Date'],
    format='mixed')
```

```python
# Check for missing values in each column
Superstore.isnull().sum()
```

```python
# Count the number of duplicate rows
Superstore.duplicated().sum()
```

### Data Cleaning Summary
Missing Values:
No missing values found in the dataset.

Duplicate Rows:
No duplicate rows found in the dataset.

Conclusion:
The dataset is clean and ready for analysis.

## Feature Engineering

New features such as Year, Month, Quarter, and Day are extracted from the Order Date to enable time-based analysis.

```python
# Extract the year from the Order Date
Superstore['Year'] = Superstore['Order Date'].dt.year
```

```python
# Extract the month name from the Order Date
Superstore['Month'] = Superstore['Order Date'].dt.month_name()
```

```python
# Extract the quarter from the Order Date
Superstore['Quarter'] = Superstore['Order Date'].dt.quarter
```

```python
# Extract the day name from the Order Date
Superstore['Day'] = Superstore['Order Date'].dt.day_name()
```

```python
# Display the newly created date-related columns
Superstore[['Order Date', 'Year', 'Month', 'Quarter', 'Day']].head()
```

# Exploratory Data Analysis (EDA)
## In this section, we analyze the Superstore dataset to answer key business questions using Pandas. The analysis focuses on sales performance, profitability, customer behavior, product performance, and regional trends.

## Key Performance Indicators (KPIs)

### Business Question 1: What is the Total Sales?


```python
total_sales = Superstore['Sales'].sum()

print(f"Total Sales: ${total_sales:,.2f}")
```

### Business Question 2: What is the Total Profit?


```python
total_profit = Superstore['Profit'].sum()

print(f"Total Profit: ${total_profit:,.2f}")
```

### Business Question 3: How many orders were placed?


```python
total_orders = Superstore['Order ID'].nunique()

print(f"Total Orders: {total_orders}")
```

### Business Question 4: What is the Average Sales?

```python
average_sales = Superstore['Sales'].mean()

print(f"Average Sales: ${average_sales:.2f}")
```

### Business Question 5: What is the Average Profit?

```python
average_profit = Superstore['Profit'].mean()

print(f"Average Profit: ${average_profit:.2f}")
```

## Category Analysis

### Business Question 6: Sales by Category

```python
sales_by_category = Superstore.groupby('Category')['Sales'].sum().sort_values(ascending=False)

print(sales_by_category)
```

### Business Question 7: Profit by Category

```python
profit_by_category = Superstore.groupby('Category')['Profit'].sum().sort_values(ascending=False)

print(profit_by_category)
```

## Regional Analysis

### Business Question 8: Sales by Region

```python
sales_by_region = Superstore.groupby('Region')['Sales'].sum().sort_values(ascending=False)

print(sales_by_region)
```

### Business Question 9: Profit by Region

```python
profit_by_region = Superstore.groupby('Region')['Profit'].sum().sort_values(ascending=False)

print(profit_by_region)
```

### Business Question 10: Sales by Segment

```python
sales_by_segment = Superstore.groupby('Segment')['Sales'].sum().sort_values(ascending=False)

print(sales_by_segment)
```

## Customer & Product Analysis

### Business Question 11: Top 10 Customers by Sales

```python
top_customers = (
    Superstore.groupby('Customer Name')['Sales']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

print(top_customers)
```

### Business Question 12: Top 10 Products by Sales

```python
top_products = (
    Superstore.groupby('Product Name')['Sales']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

print(top_products)
```

### Business Question 13: Top 10 States by Sales

```python
top_states = (
    Superstore.groupby('State')['Sales']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

print(top_states)
```

### Business Question 14: Bottom 10 States by Profit

```python
bottom_states = (
    Superstore.groupby('State')['Profit']
    .sum()
    .sort_values()
    .head(10)
)

print(bottom_states)
```

### Business Question 15: Top 10 Cities by Sales

```python
top_cities = (
    Superstore.groupby('City')['Sales']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

print(top_cities)
```


## Time Series Analysis

### Business Question 16: Sales by Year

```python
sales_by_year = (
    Superstore.groupby('Year')['Sales']
    .sum()
    .sort_values(ascending=False)
)

print(sales_by_year)
```

### Business Question 17: Profit by Year

```python
profit_by_year = (
    Superstore.groupby('Year')['Profit']
    .sum()
    .sort_values(ascending=False)
)

print(profit_by_year)
```

### Business Question 18: Sales by Month

```python
sales_by_month = (
    Superstore.groupby('Month')['Sales']
    .sum()
)

print(sales_by_month)
```

### Business Question 19: Profit by Month

```python
profit_by_month = (
    Superstore.groupby('Month')['Profit']
    .sum()
)

print(profit_by_month)
```

### Business Question 20: Sales by Ship Mode

```python
sales_by_shipmode = (
    Superstore.groupby('Ship Mode')['Sales']
    .sum()
    .sort_values(ascending=False)
)

print(sales_by_shipmode)
```

## Data Visualization

### Visualization 1: Sales by Category

```python
# Create a bar chart for sales by category

sales_by_category.plot(kind='bar', figsize=(8,5))

plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.xticks(rotation=0)

plt.show()
```

### Visualization 2: Profit by Category

```python
# Create a bar chart for profit by category

profit_by_category.plot(kind='bar', figsize=(8,5))

plt.title("Profit by Category")
plt.xlabel("Category")
plt.ylabel("Total Profit")
plt.xticks(rotation=0)

plt.show()
```

### Visualization 3: Sales by Region

```python
# Create a bar chart for sales by region

sales_by_region.plot(kind='bar', figsize=(8,5))

plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.xticks(rotation=0)

plt.show()
```

### Visualization 4: Profit by Region

```python
# Create a bar chart for profit by region

profit_by_region.plot(kind='bar', figsize=(8,5))

plt.title("Profit by Region")
plt.xlabel("Region")
plt.ylabel("Total Profit")
plt.xticks(rotation=0)

plt.show()
```

### Visualization 5: Top 10 Customers by Sales

```python
# Create a horizontal bar chart for the top 10 customers

top_customers.plot(kind='barh', figsize=(10,6))

plt.title("Top 10 Customers by Sales")
plt.xlabel("Total Sales")
plt.ylabel("Customer Name")

plt.show()
```

### Visualization 6: Top 10 Products by Sales

```python
# Create a horizontal bar chart for the top 10 products

top_products.plot(kind='barh', figsize=(10,6))

plt.title("Top 10 Products by Sales")
plt.xlabel("Total Sales")
plt.ylabel("Product Name")

plt.show()
```

### Visualization 7: Top 10 States by Sales

```python
# Create a horizontal bar chart for the top 10 states

top_states.plot(kind='barh', figsize=(10,6))

plt.title("Top 10 States by Sales")
plt.xlabel("Total Sales")
plt.ylabel("State")

plt.show()
```

### Visualization 8: Yearly Sales Trend

```python
# Create a line chart for yearly sales

sales_by_year.sort_index().plot(figsize=(8,5), marker='o')

plt.title("Yearly Sales Trend")
plt.xlabel("Year")
plt.ylabel("Total Sales")

plt.grid(True)

plt.show()
```

### Visualization 9: Yearly Profit Trend

```python
# Create a line chart for yearly profit

profit_by_year.sort_index().plot(figsize=(8,5), marker='o')

plt.title("Yearly Profit Trend")
plt.xlabel("Year")
plt.ylabel("Total Profit")

plt.grid(True)

plt.show()
```

### Visualization 10: Sales by Ship Mode

```python
# Create a bar chart for sales by ship mode

sales_by_shipmode.plot(kind='bar', figsize=(8,5))

plt.title("Sales by Ship Mode")
plt.xlabel("Ship Mode")
plt.ylabel("Total Sales")
plt.xticks(rotation=15)

plt.show()
```

# Business Insights

Based on the analysis of the Superstore dataset, the following key business insights were identified:

1. Technology was the highest-performing product category in terms of total sales.

2. The West region generated the highest overall sales, indicating strong market performance in that region.

3. The Consumer segment contributed the largest share of total sales among all customer segments.

4. A small group of customers accounted for a significant portion of total revenue, highlighting the importance of retaining high-value customers.

5. Certain products consistently generated higher sales than others, making them key revenue drivers.

6. Some states recorded negative profit despite generating sales, indicating potential issues with pricing, discounts, or operational costs.

7. Standard Class was the most frequently used shipping mode and contributed the highest sales volume.

8. Sales showed an upward trend over the years, indicating business growth.

9. Monthly sales varied throughout the year, suggesting seasonal demand patterns.

10. The analysis identified opportunities to improve profitability by focusing on high-performing products, profitable regions, and reducing losses in underperforming states.
